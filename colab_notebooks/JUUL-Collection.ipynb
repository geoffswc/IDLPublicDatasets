```python
import tempfile
import whisper
import ffmpeg
import os
import urllib
import requests
import urllib.request
import pandas as pd
```

```python
%%time
start = ['0','100','200','300']
uris = []

j = 0

session = requests.Session()
session.verify = False

for s in start:


    url = 'https://metadata.idl.ucsf.edu/solr/ltdl3/query?q=(brand:juul%20AND%20type:video)&wt=json&start=' + s
    r = session.get(url)
    data = r.json()

    for i, d in enumerate(data['response']['docs']):
        j += 1
        if 'filename' in d:
            if 'mp4' in d['filename']:
                uri = 'https://archive.org/download/tobacco_' + d['id'] + '/' + d['id'] + '.mp4'
                uris.append(uri)
```

```python
# Load the MP4 file from the URI
#video_uri = "https://archive.org/download/tobacco_wbr62a00/tobacco_wbr62a00.mp4"

# Create a temporary directory to store the downloaded MP4 file
temp_dir = tempfile.TemporaryDirectory()
```

```python
tiny_model = whisper.load_model("tiny")
#base_model = whisper.load_model("base")
#small_model = whisper.load_model("small")
#medium_model = whisper.load_model("medium")
#large_model = whisper.load_model("large")
```

```python

# Download the MP4 file to the temporary directory
#ffmpeg.input(video_uri).output(os.path.join(temp_dir.name, "video.mp4")).run()

# Load the downloaded MP4 file
url = "https://archive.org/download/tobacco_wbr62a00/tobacco_wbr62a00.mp4"
source_file_name = url.split('/')[-1]
print("downloading file")
urllib.request.urlretrieve(url, source_file_name)

source_file = os.path.join(source_file_name)

transcripts = []

for i, url in enumerate(uris[:]):#[:10]:

    try:
        source_file_name = url.split('/')[-1]
        print(i)
        print("downloading", source_file_name)
        urllib.request.urlretrieve(url, source_file_name)
        print("transcribing", source_file_name)

        source_file = os.path.join(source_file_name)
        result = tiny_model.transcribe(source_file)
        transcripts.append((url, result['text']))
        print('removing', source_file)
        os.remove(os.getcwd() + '/' + source_file)
        print()
    except:
        print("failed", source_file)
        transcripts.append((url, 'None'))
```
