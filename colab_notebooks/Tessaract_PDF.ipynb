```python
import pytesseract
import tempfile
import urllib
import os
from PIL import Image
from pdf2image import convert_from_path
```

```python
#Create a temporary directory to store the downloaded pdf file
temp_dir = tempfile.TemporaryDirectory()

url = "https://download.industrydocuments.ucsf.edu/z/k/j/p/zkjp0185/zkjp0185.pdf"

source_file_name = url.split('/')[-1]
print("downloading file")
urllib.request.urlretrieve(url, source_file_name)
source_file = os.path.join(source_file_name)
print(source_file)
```

```python
%time
imgs = convert_from_path(source_file,jpegopt=True)
```

```python
%%time
transcriptions = []
str_out = ""
for img in imgs[:5]:
  print(img)
  transcriptions.append(pytesseract.image_to_string(img))
```
