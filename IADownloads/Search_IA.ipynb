To find these types of videos:
Go to UCSF Industry Archives Videos page on Internet Archive

Depositions: search “depositions” (144 results: https://archive.org/details/industry-archives?query=deposition )
    
Corporate Communications: search “meeting OR training OR employee” (458 results: https://archive.org/details/industry-archives?query=meeting+OR+training+OR+employee)

Focus Group Interviews: search “focus group” (146 results: https://archive.org/details/industry-archives?query=%22focus+group%22 )





```python
from internetarchive import search_items
from internetarchive import download

# parsing a collection
# https://archive.org/services/docs/api/items.html?highlight=collection
#for item in search_items('identifier:tobacco_stkd0111').iter_as_items():
#    print('item', item)
#    print(item.item_metadata)

#for item in search_items('tobacco').iter_as_items():
#    i += 1
    #print('item', item)
    #print(item.item_metadata)

#items = search_items('collection:industry-archives AND (meeting OR training OR employee)')
#458 results

items = search_items('collection:industry-archives AND (deposition)')
#144 results

#items = search_items('collection:industry-archives AND (focus group)')
#151 results - note this one is slightly different

#UCSF Industry Archives Videos

print(len(items))

# 'industry-archives'
i = 0

max_download = 500
for item in items.iter_as_items():
    i += 1
    #print(i, 'item', item)
    #print(item.item_metadata['files'])
    identifier = item.item_metadata['metadata']['identifier']
    #download(identifier, verbose=True)
    #print(identifier)

    download(identifier, verbose=True, glob_pattern='*xml')
    
    if i > max_download:
        break
    
```
